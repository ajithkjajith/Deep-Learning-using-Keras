Loading Dataset

```
import numpy as np
import pandas as pd

dataset = pd.read_csv("concrete_data.csv")
dataset.head()
```

```
import keras
```

```
from keras.models import Sequential # for linear stack of inputs
from keras.layers import Dense # for layers in network
```

Dividing predictors and target as x and y

```
dataset_columns = dataset.columns

x = dataset[dataset_columns[dataset_columns != 'Strength']] # all columns except Strength
y = dataset['Strength'] # Strength column
```

- One hidden layer of 10 nodes, and a ReLU activation function

- Use the adam optimizer and the mean squared error as the loss function.

```
model = Sequential()
model.add(Dense(10, activation='relu', input_shape=(8,)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')
```

Splitting training and testing dataset

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=0)
```

```
# fit the model
model.fit(X_train,y_train, epochs=50, verbose=1)
```

```
#finding prediction from model built
y_pred = model.predict(X_test)
```

```
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test,y_pred)
```

```
print("Mean = ",np.mean(mse))
print("Standard Deviation = ",np.std(mse))
```

  For 50 squared errors
  

```
mean_squared_errors=[]
for i in range(0,50):
  X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
  model.fit(X_train,y_train, epochs=50, verbose=1)
  y_pred = model.predict(X_test)
  mean_square_error = mean_squared_error(y_test, y_pred)
  mean_squared_errors.append(mean_square_error)
```

```
mean_squared_errors = np.array(mean_squared_errors)
mean = np.mean(mean_squared_errors)
standard_deviation = np.std(mean_squared_errors)

print("Mean for 50 mean squared errors is ",mean)
print("Standard deviation for 50 mean squared errors is ",standard_deviation)
```
